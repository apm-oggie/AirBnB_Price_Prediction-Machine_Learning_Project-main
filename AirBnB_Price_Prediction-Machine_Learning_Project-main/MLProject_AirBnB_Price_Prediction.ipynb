***MODEL TRAINING , EVALUATION , SELECTION***

```python
#Tranining and Testing data splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
import xgboost as xgb
lr=LinearRegression()
tree_reg = DecisionTreeRegressor()
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
xgbr = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
lst=[lr,tree_reg,forest_reg,gbr,xgbr]
```

```python
#Performance Evaluation
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
for i in lst:
  print(" ")
  print("Model name is",i)
  print("************************************************************************")
  i.fit(x_train,y_train)
  print("prediction value is : ")
  y_pred=i.predict(x_test)
  print(y_pred)
  print(" ")
  print("Mean Absolute Error is : ",mean_absolute_error(y_test,y_pred))
  print("Mean Absolute Percentage Error  is : ",mean_absolute_percentage_error(y_test,y_pred))
  print("Mean Squared Error is : ",mean_squared_error(y_test,y_pred))
  print("R2 Score is : ",r2_score(y_test,y_pred))
```

***INSIGHTS FROM THE PERFORMANCE METRICS OF THE DIFFERENT REGRESSION MODELS USED***




*  All models show high R2 scores, indicating that they all fit the data well.
*   **Linear Regression** has the lowest MAE and MAPE, suggesting it might be the best performing model in terms of error rate.


*   **Decision Tree Regressor** and **Random Forest Regressor** offer comparable performance, with Random Forest showing slightly better metrics.
*   **Gradient Boosting Regressor** and **XGBoost Regressor** have higher error rates but still maintain strong predictive capabilities as indicated by their R2 scores.
--------------------
In conclusion, while Linear Regression provides the best error metrics, Decision Tree and Random Forest Regressors offer robust performance with potentially better generalizability due to their ensemble nature. Gradient Boosting and XGBoost are also strong contenders, particularly for datasets where boosting might provide advantages despite higher computational cost.
